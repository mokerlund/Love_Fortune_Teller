```python
import pandas as pd

raw_file_df = pd.read_csv('CSV_files/HCMST_ver_3.csv')
raw_file_df.head()
```

```python
# View all columns

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
raw_file_df.columns.values
```

```python
# Count of partnered vs not

raw_file_df["qflag"].value_counts()
```

### Notes:

* Remove all non partnered applicants
* Remove all with w2_ and w3_ prefixes
* Remove all demographic info
* Keep Q18 -Q35

```python
# Filter for only applicants in relationship

partnered_only_df = raw_file_df[raw_file_df["qflag"] == 'partnered']
partnered_only_df["qflag"].value_counts()
```

```python
# Drop columns that start with w2 or w3 to keep only the first wave
# Drop all demographic info

wave_1_df = partnered_only_df.drop(columns=partnered_only_df.iloc[:, 2:51])
wave_1_df = wave_1_df.loc[:,~wave_1_df.columns.str.startswith('w3')]
wave_1_df = wave_1_df.loc[:,~wave_1_df.columns.str.startswith('w2')]
wave_1_df = wave_1_df.loc[:,~wave_1_df.columns.str.startswith('pp')]
wave_1_df = wave_1_df.drop(columns=['interstate_mover_pp1_pp2', 'interstate_mover_pp1_pp3', 'interstate_mover_pp2_pp3', 'Unnamed: 0'])

# Reset Index
wave_1_df = wave_1_df.reset_index(drop=True)

wave_1_df.head()
```

```python
# dictionary = wave_1_df.columns.values
# dictionary_df = pd.DataFrame(dictionary)
wave_1_df.to_csv('wave_1_main_questions.csv', index=False)
```

```python
# Export to sqlite database

from sqlalchemy import create_engine

# Empty connection until all tables will be comined in a single database
engine = create_engine('sqlite:///Relationship_survey.sqlite', echo=False)

wave_1_df.to_sql('relationship_qs', con=engine, if_exists='replace')
```
