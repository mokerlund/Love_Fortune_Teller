```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
raw_df = pd.read_csv("project_3_raw_data_wv1.csv")
```

```python
raw_df.head()
```

```python
larger_region = raw_df.ppmsacat
larger_region.value_counts()
```

```python
selected = raw_df[["caseid_new", "q21a", "q16", "ppreg4", "ppmsacat"]]
```

```python
selected.head()
```

```python
unclean_features = selected.rename(columns={"q21a": "age_when_met", "q16":"family_size", "ppreg4":"region", "ppmsacat":"metro"})
```

```python
unclean_features.region.value_counts()
```

```python
unclean_features.head()
```

```python
unclean_features = pd.concat([unclean_features, pd.get_dummies(unclean_features["region"], prefix="region")], axis=1)
unclean_features['region'] = unclean_features.region.map({'south':0, 'west':1, 'midwest':2, 'northeast':3})
unclean_features['metro'] = unclean_features.metro.map({'non-metro':0, 'metro':1})

# region map key: 'south':0, 'west':1, 'midwest':2, 'northeast':3

df = unclean_features
df
```

```python
df.to_csv("age_family_region_clean.csv")
```
